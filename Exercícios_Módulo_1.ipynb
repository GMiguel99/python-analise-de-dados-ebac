# Exercício Números

# Exercício String

## Parte 1

```python
# Exercício 1
cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'

# Método 1 - Upper
upper_cancao = cancao.upper()
print(upper_cancao, "\n")

# Método 2 - Find
posicao = cancao.find("mundo")
print(posicao, "\n")

# Método 3 - Replace
nova_cancao = cancao.replace("roda", "1234")
print(nova_cancao)
```

## Parte 2

```python
noticia = 'Selic vai a 2,75% e supera expectativas; ' + 'é a primeira alta em 6 anos.'

selic = noticia[12:16]
ano = noticia[-7]
print(selic, "|", ano)
```
